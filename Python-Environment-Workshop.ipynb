# Virtual environments and how to manage them

As discussed earlier we need environments to work efficiently. But why do we need one? Say you want to do a certain task and it requires a set of libraries but you don't need them in another project. In that case, we create 2 environments with libraries specific to one project in 1 environment and some other libraries specific to another project in another environment. This gives us the power to keep things independent yet closely related to each other.

### 1. Conda Environment creation

To create a conda environment, run the following command:

```bash
conda create --name py3.6
```

The above line will create an environment by the name py3.6. You can similarly create multiple environments. To see a list of environments on your machine, run:

```bash
conda info --envs
```
or

```bash
conda env list
```

For more information on [Conda Environments](https://conda.io/docs/user-guide/tasks/manage-environments.html)

### 2. Activating and deactivating a conda environment

To activate a conda environment, if on Windows type:

```bash
activate py3.6
```

else if on Linux or MAC, type:

```bash
source activate py3.6
```

The above command will activate your environment and you will see name of your in the prompt. Here whatever libraries you install will be specific to that environment itself. 

To deactivate a conda environment, for windows just run: 
```bash
deactivate
```

For Linux and MAC, run: 
```bash
source deactivate
```


# Installing libraries and packages

### INSTALLING CONDA PACKAGES:

**To list and check for all of the packages in the active environment:**

conda list

**To check all the available environments:**

SET (in cmd)

**To list all of the packages in a deactivated environment:**

conda list -n env_path

**To see if a specific package is available for installation:**

conda search Package_Name

**To see if a specific package is available for installation from Anaconda.org:**

conda search --override-channels --channel defaults Package_Name

**To install a specific package into an existing environment:**

conda install --name env_path Package_Name

**If you do not specify the environment name, which in this example is done by --name <env_path>, the package installs into the current environment:**

conda install Package_Name

**To install a specific version of a package:**

conda install Package_Name=0.15.0

**To install multiple packages at once:**

conda install PackageName1 PackageName2

NOTE: It is best to install all packages at once, so that all of the dependencies are installed at the same time.

**To install multiple packages at once and specify the version of the package:**

conda install PackageName1=0.15.0 PackageName2=7.26.0

**To update a specific package:**

conda update Package_Name

**To update Python:**

conda update python

**To update conda itself:**

conda update conda

NOTE: Conda updates to the highest version in its series, so Python 2.7 updates to the highest available in the 2.x series and 3.6 updates to the highest available in the 3.x series.

**To remove a package in an environment:**

conda remove -n env_path Package_Name

**To remove a package in the current environment:**

conda remove Package_Name

**To remove multiple packages at once:**

conda remove PackageName1 PackageName2

**To confirm that a package has been removed:**

conda list


### INSTALLING NON-CONDA PACKAGES :

If a package is not available from conda or Anaconda.org, you may be able to find and install the package with another package manager like pip.

Pip packages do not have all the features of conda packages, and we recommend first trying to install any package with conda. If the package is unavailable through conda, try installing it with pip. The differences between pip and conda packages cause certain unavoidable limits in compatibility, but conda works hard to be as compatible with pip as possible.

NOTE: Both pip and conda are included in Anaconda and Miniconda, so you do not need to install them separately.

NOTE: Conda environments replace virtualenv, so there is no need to activate a virtualenv before using pip.

It is possible to have pip installed outside a conda environment or inside a conda environment.

To gain the benefits of conda integration, be sure to install pip inside the currently active conda environment, and then install packages with that instance of pip. The command 'conda list' shows packages installed this way, with a label showing that they were installed with pip.

You can install pip in the current conda environment with the command 'conda install pip'.

If there are instances of pip installed both inside and outside the current conda environment, the instance of pip installed inside the current conda environment is used.

**To install a non-conda package:**

Activate the environment where you want to put the program:

1. On Windows, in your Anaconda Prompt, run activate <env_path>.
2. On macOS and Linux,in your Terminal window, run source activate <env_path>.

To use pip to install a program such as See, in your Terminal window or an Anaconda Prompt, run:

pip install see

To verify the package was installed, in your Terminal window or an Anaconda Prompt, run:

conda list

NOTE: If the package is not shown, install pip and try these commands again.

**Upgrading pip**

conda update pip

**Checking version of pip**

pip --version

**Installing a package**

pip install Package_Name

**Check for the files installed for a package**

pip show --files Package_Name
    
**List installed packages**

pip list

**List uptodate packages**

pip list --uptodate

**Check for outdated files**

pip list --outdated

**List packages that are not dependencies of other packages. Can be combined with other options**

pip list --outdated --not-required

**Upgrading a package**

pip install --upgrade Package_Name
    
**Uninstalling a package**

pip uninstall Package_Name

# A Dive into Jupyter Notebooks

Jupyter Notebook is a web application that allows you to create and share documents that contain:

1. live code (e.g. Python code)
2. visualizations
3. explanatory text (written in markdown syntax)

Jupyter Notebook is great for the following use cases:

1. practise Python
2. data processing / transformation
3. numeric simulation
4. statistical modeling
5. machine learning

The user interface of Jupyter Notebook is split up into three sections (tabs):

1. Files - The default view is the Files tab from where you can open or create notebooks.
2. Running - Running tab shows you all the notebooks and terminals which are running on Jupyter.
3. Clusters - Clusters tab is provided by IPython parallel. See 'IPython parallel' for installation details from the Jupyter Notebook.

**Creating a new Jupyter Notebook:**

Creating a new Jupyter Notebook is easy. Just use the New dropdown menu and select Python3 to open a new Jupyter Notebook for Python. The notebook is created but untitled. By clicking into the text “Untitled” on the top you can give it a name. By giving it a name the notebook will also be saved as a file of the same name with extension .ipynb. E.g. name the notebook myFirstNotebook.

**Working With The Notebook:**

The notebook itself consists of cells. A first empty cell is already available after having created the new notebook.This cell is of type “Code” and you can start typing in Python code directly. Executing code in this cell can be done by either clicking on the run cell button or hitting Shift + Return keys. The resulting output becomes visible right underneath the cell. The next empty code cell is created automatically and you can continue to add further code to that cell.

You can change the cell type from Code to Markdown to include explanatory text in your notebook. Once switched the type to Markdown you can start typing in markdown code. After having entered the markdown code you can compile the cell by hitting Shift + Enter/Return once again. The markdown editor cell is then replaced with the output. If you want to change the markdown code again you can simply click into the compiled result and the editor mode opens again.

**Edit And Command Mode:**

If a cell is active two modes distinguished:
1. edit mode
2. command mode

If you just click in one cell, the cell is opened in command mode which is indicated by a blue border on the left.
If you click into the code area of the cell, the edit mode is enabled which is indicated by a green border on the left.
If you’d like to leave edit mode and return to command mode again you just need to hit ESC.

**Checkpoints:**

Another function of Jupyter Notebook is the ability to create checkpoint. By creating a checkpoint you’re storing the current state of the notebook so that you can later on go back to this checkpoint and revert changes which have been made to the notebook in the meantime.

To create a new checkpoint for your notebook select menu item Save and Checkpoint from the File menu. The checkpoint is created and the notebook file is saved. If you want to go back to that checkpoint at a later point in time you need to select the corresponding checkpoint entry from menu File → Revert to Checkpoint.

**Exporting The Notebook:**

Jupyter Notebook gives you several options to export your notebook. Those options can be found in menu File → Download as.

# Version Control with Github

Version control with github is very important and hence its necessary to know some commands.

### Part 1: To push your changes to github

``` git
git status

git add *filename* 
```

use above snippet if adding files for the first time else do
```git
git add -u 
```

for pushing changed files to staging area from the unstaged area

```git
git commit -m *commit_message*

git push
```


### Part 2: Branch Creation

By default we are on master. Its a good practice to have seperate branch for development and other feature development.

To create a new branch from master, do:
```git
git checkout -b *branch_name*
```

This will create a new branch of that name you mentioned and will take in all the codes which may be present in the master. This branch is not tracked on remote, hence you need to create an upstream for it. You can do it by typing:

```git
git push --set-upstream origin *branch_name_on_which_you_are_there*
```

Its a good practice to have branches with the following name created from master at the very beginning.
develop - a copy of master. if things work good here they are elligible to be pushed to master
sandbox - a test branch for unit tests or testing any feature or any other release or bug fix. It is optional since you will be having tests on your own branch as well, yet a good practice.
*your_name* - branch were you work on your part of the code. This is the place where you develop your code.

*your_name* is branch specific to the developer. When working in a team and on a same software project, it becomes easier to work on your code which is on a branch private to just you.

For more about [Branching](https://www.atlassian.com/git/tutorials/using-branches)

### Part 3: Branch Merging

When you have finished working on your code, you need to incorporate those changes on the branch common to all users, in our case it can be either **sandbox** for testing or the **develop** for final pushes.

For branch merging, always remember the following rules - 

1. Commit and push the changes on your branch as described in Part 1
2. Pull the latest changes from the develop branch (assuming you will be merging your branch to develop) using following command: 
```git
git pull origin develop
```
3. Push those new changes to your branch having resolved the conflicts
4. Checkout to the develop branch using: 
```git
git checkout develop
```
5. Pull your branch to merge into develop.the below code will merge your branch into develop: 
```git
git pull origin *your_branch*
```

6. Finally push the updated develop using : 
```git 
git push
```
Alternate Method

Alternatively, when you have commited the changes and pushed it on your branch, you can checkout to the branch where you want to merge your current branch. Say you want to merge your current branch **sandbox** to **develop**. So having commited changes to sandbox, you can checkout to develop and then do the following:

```git
git merge sandbox
```
And then push it using git push.

To summarize, always follow the above procedure for merging into any branch. The develop can be replaced by any branch into which you would like to merge.

For more information about [Merging](https://www.atlassian.com/git/tutorials/using-branches/git-merge)

### Part 4: Stashing 

Let's say there is an important piece of code which needs to be incorporated into your branch and you still haven't finished your work, or your friend is stuck somewhere and you need to go to his branch to solve the issue and here too, your work is not finished, in such cases, you cannot push your incomplete code. Stashing helps you save the current state of your code in memory so that when you come back to your branch of have incorporated some external changes, you can retrieve the saved changes back into your code. 

To stash use: 
```git
git stash
```

When you return to your branch again or have completed incorporating external changes do: 
```git
git stash pop 
```
This will reapply the stashed changes to your code and remove it from the memory stack.

For more about [Stashing](https://www.atlassian.com/git/tutorials/saving-changes/git-stash)

### Part 5: Resolving Conflicts

Every now and then, there might arise situations and cases where you have to resolve merge conflicts. Merge conflicts arise when 2 version of the same part of the code exist and git does not know what to do. It either resolves most of them automatically, but in some places it asks for user intervention for resolving the conflicts.

Conflicts look somewhat like this
```
<<<<<<< HEAD
print(a + b)
=======
print(a * b).
>>>>>>> branch-name
```

In such case, git does not know if you want code **print(a + b)** or **print(a * b)** to be kept. So the part above ==== is something which is there on the local currently, while the part below ==== is something which came from the remote location to your branch either due to merge or pull. You should keep either one of the code and remove *<<<<< HEAD*, *=====* and *>>>>>> branch-name* lines.

Finally, commit your changes and push it.

For more information on [Conflict Resolution](https://www.atlassian.com/git/tutorials/using-branches/merge-conflicts)

### Part 6: Branch Deletion

Often at times when you have to get some bugs fixed or feature incorporated, you may create a seperate branch for development. At times, too many branches can make your development lifecycle quite congested; so a good practice is to delete unwanted or dormant branches. 

Deleting a branch is a 2 way process since you need to remove both the local as well as the remote branches. 

To remove a local branch you should **NOT BE ON THAT BRANCH**. Well, this is quite obvious since you can't remove a branch, while you are on it. So as a first step, checkout to master branch using:

```git
git checkout master
```

Next, to delete a local branch, say sandbox, run

```git
git branch -d sandbox
```
The above line removes your branch locally, but it still exists on remote, so to delete it completely, run:

```git 
git push origin --delete sandbox
```
For more on [Branch Deletion](https://www.atlassian.com/git/tutorials/using-branches)

### Part 7: Configuring .gitignore file

You certainly don't want to push all the files to github. Many a times you have data files, or .ipynb_checkpoints which needs to be ignored to be pushed to github. So basically .gitignore files are not tracked by git and they remain on your local machine itself.

If .gitignore file does not exist, create one.

Then place the files which you want to ignore. Say you want to ignore, .ipynb_checkpoints everywhere in your directory, then you place a line in .gitignore as:

*.ipynb_checkpoints

This helps you ignore all the files where the extension is .ipynb_checkpoints and it is not tracked by git at all. 

For more on [.gitignore](https://www.atlassian.com/git/tutorials/saving-changes/gitignore)

### Part 8: Configuring README.md

Readme is a file which explains your project. It has information on project description, installation steps, libraries used, citations, authors... etc. Readme is written in Markdown language. If you don't want to get into the nuts and bolts of HTML and CSS, then Markdown helps you create web elements. A better way to visualize it is, it acts like a wrapper.

For the cheat-sheet on [Markdown](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet)

Also recall, markdown is integrated in Jupyter Notebooks as seen earlier.

For a template on [README.md](https://gist.github.com/PurpleBooth/109311bb0361f32d87a2)

# Development Hacks

**Keyboard Shortcuts for Jupyter Notebook:**

Press ESC and below keys for shortcuts in Windows -

1. A - to insert cell above
2. B - to insert cell below
3. DD - to delete selected cells
4. F - to find and replace
5. H - to show keyboard shortcuts
6. I - to interrupt the kernel
7. J - to select cell below
8. K - to select cell above
9. L - to toggle line numbers
10. M - to change cell to markdown
11. O - to toggle output of selected cells
12. S/ Ctrl-S - to save and checkpoint
13. Up - to select cell above
14. Down - to select cell below

**Some PyCharm Hacks**

1. Splitting the Window - Right clicking on a code tab will give you options to split windows either vertically / horizontally.
2. Deleting line of code - Ctrl + Y (Windows and Linux), Cmd + Y (MAC)
3. Reformating code (PEP8 style): Ctrl + Alt + L (Windows and Linux), Option + Cmd + L(MAC)
4. GLobal search: Double tap Shift
5. Fold / unfold functions: Ctrl + Shift + (+ or -) (windows and Linux), Cmd + Shift + (+ or -) (MAC)
6. Function documentation: triple quotes after function definition.

## CONCLUSION

This workshop has helped you get started with your data science journey. It assists in setting up the environment, installation of Anaconda and Pycharm, installation of various python libraries with conda and pip and understanding the working of Jupyter Notebook and GIT.

# License


<div class="alert alert-block alert-info">
Copyright (c) 2019 Nikunj Lad
<br><br>
Permission is hereby granted, free of charge, to any person obtaining a copy of this software and associated documentation files (the "Software"), to deal in the Software without restriction, including without limitation the rights to use, copy, modify, merge, publish, distribute, sublicense, and/or sell copies of the Software, and to permit persons to whom the Software is furnished to do so, subject to the following conditions:
<br><br>
The above copyright notice and this permission notice shall be included in all copies or substantial portions of the Software.
<br><br>
THE SOFTWARE IS PROVIDED "AS IS", WITHOUT WARRANTY OF ANY KIND, EXPRESS OR IMPLIED, INCLUDING BUT NOT LIMITED TO THE WARRANTIES OF MERCHANTABILITY, FITNESS FOR A PARTICULAR PURPOSE AND NONINFRINGEMENT. IN NO EVENT SHALL THE AUTHORS OR COPYRIGHT HOLDERS BE LIABLE FOR ANY CLAIM, DAMAGES OR OTHER LIABILITY, WHETHER IN AN ACTION OF CONTRACT, TORT OR OTHERWISE, ARISING FROM, OUT OF OR IN CONNECTION WITH THE SOFTWARE OR THE USE OR OTHER DEALINGS IN THE SOFTWARE.
</div>
